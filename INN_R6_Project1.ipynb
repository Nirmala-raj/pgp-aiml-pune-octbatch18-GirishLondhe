**Geography and Gender attributes are catagorical. So we will use Label encoding to convert them.**

```
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
bankdf['Geography'] = le.fit_transform(bankdf['Geography'])
bankdf['Gender'] = le.fit_transform(bankdf['Gender'])
```

```
bankdf.head()
```

```
Target=pd.DataFrame(bankdf['Exited'])
Feature=bankdf.drop(['Exited'],axis=1)
```

```
Target.shape
```

```
Feature.shape
```

```
#Build Train / Test dataset

from sklearn.model_selection import train_test_split


X_train,X_test,y_train,y_test = train_test_split(Feature,Target,test_size=0.3)
```

**Model Building**

```
#Normalize Test and Train feature datasets


X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

# Feature Scaling 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

#Normalize
from sklearn.preprocessing import Normalizer
transformer = Normalizer()
X_train = transformer.fit_transform(X_train)
X_test = transformer.transform(X_test)
```

```

y_train['Exited'].value_counts()
```

```
#Initialize Sequential model
model = tf.keras.models.Sequential()

#Add Dense layer for prediction - Keras declares weights and bias automatically
model.add(tf.keras.layers.Dense(100, input_shape=(10,),activation='relu'))

#Add 1st hidden layer
model.add(tf.keras.layers.Dense(80, kernel_initializer='normal',activation='relu'))

#Add 2nd hidden layer
model.add(tf.keras.layers.Dense(50, kernel_initializer='normal',activation='relu'))

#Add 3rd hidden layer
model.add(tf.keras.layers.Dense(30, kernel_initializer='normal',activation='relu'))
                                
#Add 4rd hidden layer
model.add(tf.keras.layers.Dense(15, kernel_initializer='normal',activation='relu'))

#Add 4rd hidden layer
model.add(tf.keras.layers.Dense(5, kernel_initializer='normal',activation='relu'))

#Add OUTPUT layer
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
```

```
#Create optimizer with non-default learning rate
#sgd_optimizer = tf.keras.optimizers.SGD(lr=0.03)

#Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```
model.fit(X_train,y_train,          
          validation_data=(X_test,y_test),
          epochs=15)
```

```
#Optimize model

#add batching

model.fit(X_train,y_train,          
          validation_data=(X_test,y_test),
          epochs=30,batch_size=50)
```

```
y_pred=model.predict(X_test)
y_pred
```
